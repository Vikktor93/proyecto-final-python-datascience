### **CREACIÓN DE NUEVAS VARIABLES**

Como el dataset es de videouegos, una de las alternativas a considerar es la proporción de ventas por región con respecto a las ventas globales. En esta etapa se crean cuatro nuevas variables que representen las proporciones de ventas en cada región en relación con las ventas globales. Estas variables van a permitir entender de mejor forma, la distribución de las ventas en diferentes regiones.

##### **¿QUÉ EJEMPLOS DE USO PUEDE TENER ESTAS NUEVAS VARIABLES?**

Se pueden por ejemplo, utilizar para identificar patrones de preferencia regional o analizar cómo se distribuyen las ventas en diferentes mercados

##### **DESCRIPCIÓN DE LAS NUEVAS COLUMNAS**

- `NA_Sales_Ratio`: Proporción de ventas en América del Norte respecto a las ventas globales
- `EU_Sales_Ratio`:Proporción de ventas en Europa respecto a las ventas globales
- `JP_Sales_Ratio`: Proporción de ventas en Japón respecto a las ventas globales
- `Other_Sales_Ratio`: Proporción de ventas en otras regiones respecto a las ventas globales

##### **CREACIÓN DE VARIABLES REGIONALES PROPORCIONALES A LAS VENTAS GLOBALES**


```python
df['NA_Sales_Ratio'] = df['NA_Sales'] / df['Global_Sales']
df['EU_Sales_Ratio'] = df['EU_Sales'] / df['Global_Sales']
df['JP_Sales_Ratio'] = df['JP_Sales'] / df['Global_Sales']
df['Other_Sales_Ratio'] = df['Other_Sales'] / df['Global_Sales']
```
